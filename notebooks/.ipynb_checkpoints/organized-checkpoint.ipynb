$$ avgcharactes = numcomments_{post} * \sum^{comments} numcharacters $$

```python
mean_comment_size_by_post = {}
mean_comment_size_by_post['fb_search_posts'] = [np.mean([len(c['comment_text']) for c in fb_search[post][0]['comments_full']])
     for post in search_posts if fb_search[post][0]['comments_full']]
mean_comment_size_by_post['fb_search_pages'] = [np.mean([len(c['comment_text']) for c in post['comments_full']])
     for page in search_pages for post in fb_search[page] if post['comments_full']]


mean_comment_size_by_post['netmums'] = [np.mean([len(post['body']) for post in netmums[key]['posts']]) for key in nmkeys]


mean_comment_size_by_post['groups'] = [np.mean([len(c['comment_text']) for c in post['comments_full']]) for group in groups.values() for post in group if post['comments_full']]
mean_comment_size_by_post['pages'] = [np.mean([len(c['comment_text']) for c in post['comments_full']]) for page in pages.values() for post in page if post['comments_full']]
```

```python
for key in mean_comment_size_by_post.keys():
    plt.title('mean comment length per post: ' + str(key))
    plt.yscale('symlog')
    plt.plot(np.sort(mean_comment_size_by_post[key]))
    plt.show()
```

```python
for key in mean_comment_size_by_post.keys():
    plt.title('histogram: mean comment length per post: ' + str(key))
    plt.yscale('log')
    plt.hist(mean_comment_size_by_post[key], bins = 200)
    plt.show()
```

```python
comment_length_data = {}
#return a list of comment lengths, each list is a post
mean_comment_size_by_post['fb_search_posts'] = [[len(c['comment_text']) for c in fb_search[post][0]['comments_full']]
     for post in search_posts if fb_search[post][0]['comments_full']]
mean_comment_size_by_post['fb_search_pages'] = [[len(c['comment_text']) for c in post['comments_full']]
     for page in search_pages for post in fb_search[page] if post['comments_full']]


mean_comment_size_by_post['netmums'] = [[len(post['body']) for post in netmums[key]['posts']] for key in nmkeys]


mean_comment_size_by_post['groups'] = [[len(c['comment_text']) for c in post['comments_full']] for group in groups.values() for post in group if post['comments_full']]
mean_comment_size_by_post['pages'] = [[len(c['comment_text']) for c in post['comments_full']] for page in pages.values() for post in page if post['comments_full']]
```

```python
for key in mean_comment_size_by_post.keys():
    plt.title('comment length per post(quantiles): ' + str(key))
    plt.yscale('symlog')
    temp = mean_comment_size_by_post[key]
    #for each list of numbers in list get quantiles
    qs = [0.001, 0.05,0.25,0.5,0.75, 0.95, 0.999]
    temp = [np.quantile(i, qs) for i in temp]
    ind_mean = np.argsort([np.mean(qs) for qs in temp])
    ind_top = np.argsort([qs[-1] for qs in temp])
    ind_bot = np.argsort([qs[3] for qs in temp])
    for i, quantile in enumerate(qs):
        quantile_data = np.array([obs[i] for obs in temp])
        plt.plot(quantile_data[ind_bot], label=str(quantile))
    plt.legend()
    plt.show()
```

```python
for key in mean_comment_size_by_post.keys():
    plt.title('comment length per post(quantiles): ' + str(key))
    plt.yscale('symlog')
    temp = mean_comment_size_by_post[key]
    #for each list of numbers in list get quantiles
    qs = [0.001, 0.05,0.25,0.5,0.75, 0.95, 0.999]
    temp = [np.quantile(i, qs) for i in temp]

    for i, quantile in enumerate(qs):
        quantile_data = [obs[i] for obs in temp]
        plt.plot(np.sort(quantile_data), label=str(quantile))
    plt.legend()
    plt.show()
```
